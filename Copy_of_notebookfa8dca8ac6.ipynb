```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('/content/Churn_Modelling.csv')
```

```python
dataset.head()
```

#Data Cleaning

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
dup = dataset.duplicated()
dataset[dup].shape[0]
```

```python
dataset.dtypes
```

```python
dataset.describe()
```

```python
columns_to_drop = ['RowNumber', 'CustomerId', 'Surname']
existing_columns = [col for col in columns_to_drop if col in dataset.columns]
dataset.drop(columns = existing_columns, axis = 1, inplace = True)
dataset.head()
```

###Label Encoding on Geography column

```python
dataset['Geography'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['Geography'] = le.fit_transform(dataset['Geography'])
dataset['Geography']
```

###One Hot Encoding on Gender column

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ar = ohe.fit_transform(dataset[['Gender']]).toarray()
dataset['Gender'] = ar[:,0]
dataset['Gender']
```

##Seperate Feature and Target Column

```python
feature_col = dataset.iloc[:,:-1]
target_col = dataset.iloc[:,-1]
```

###Feature Scaling

```python
ss = StandardScaler()
feature_col = pd.DataFrame(ss.fit_transform(feature_col), columns=feature_col.columns)
```

##Seperate Training and Testing data

```python
x_train, x_test, y_train, y_test =  train_test_split(feature_col,target_col,test_size=0.2, random_state=42)
```

#ANN

```python
feature_col.shape
```

```python
x_test
```

```python
import tensorflow
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann = Sequential()
```

```python
ann.add(Dense(8, input_dim = 10, activation="relu"))
ann.add(Dense(6, activation="relu"))
ann.add(Dense(4, activation="relu"))
ann.add(Dense(2, activation="relu"))
ann.add(Dense(1, activation="sigmoid"))
```

```python
ann.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
ann.fit(x_train, y_train, batch_size=50, epochs=35)
```

###Evaluation

####1 - Check Testing accuracy

```python
prd = ann.predict(x_test)
prd_data =[]
for i in prd:
  if i>0.5:
    prd_data.append(1)
  else:
    prd_data.append(0)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prd_data) * 100
```

####2 - Check Training accuracy

```python
prd_data_train =[]
prd_train = ann.predict(x_train)
for i in prd_train:
  if i>0.5:
    prd_data_train.append(1)
  else:
    prd_data_train.append(0)
```

```python
accuracy_score(y_train,prd_data_train) * 100
```
