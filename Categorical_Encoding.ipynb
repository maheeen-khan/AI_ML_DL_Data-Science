```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('/content/Salaries.csv')
dataset.head()
```

```python
dataset['rank'].unique() # check classes in rank column
```

###Label Encoding on 'rank' column

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder() # make instance of LabelEncoder object
```

```python
dataset['rank'] = le.fit_transform(dataset['rank']) # training and applying encoding
```

```python
dataset['rank'] # check data
```

```python
dataset['rank'].unique() # after encoding, check unique classes
```

After label encoding:


*    0 assigns to Associate Professor,
*    1 assigns to Assistant Professor,
*    2 assigns to Professor




### One Hot Encoding by Pandas

```python
df = pd.DataFrame(
    {
        "colors" : ["red", "green", "blue", "red", "green", "blue", "red"]
    }
)

df
```

```python
oneHot = pd.get_dummies(df,columns=['colors'])
oneHot
```

### One Hot Encoding by Sklearn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False)
```

```python
arr = ohe.fit_transform(df[['colors']])
arr
```

```python
pd.DataFrame(arr,columns=['colors_red','colors_green','colors_blue'])
```

### Ordinal Encoding by Sklearn

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['A','B']]) # its mean A -> 0 , B -> 1
```

```python
dataset['discipline'] = oe.fit_transform(dataset[['discipline']])
```
